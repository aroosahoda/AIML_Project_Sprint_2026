```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("data/Combined12.csv")

df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df["temp_mean(c)"].hist()
plt.title("Temperature Distribution")
plt.show()
```

```python
plt.scatter(df["temp_mean(c)"], df["global_radiation"])
plt.xlabel("temp_mean(c)")
plt.ylabel("global_radiation")
plt.title("Temp vs Radiation")
plt.show()
```

```python
sns.boxplot(x=df["Pressure"])
plt.title("Pressure Boxplot")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
